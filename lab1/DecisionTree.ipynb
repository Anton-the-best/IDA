```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
pistachio = pd.read_csv("pistachio.csv") #https://www.kaggle.com/datasets/amirhosseinmirzaie/pistachio-types-detection?resource=download
pistachio["Class"] = pistachio["Class"].replace({"Kirmizi_Pistachio":0,"Siit_Pistachio":1})
pistachio
```

```python
X = pistachio.drop(['Class'], axis = 1)
y = pistachio['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
scaled_X_train = scaler.transform(X_train) 
scaled_X_test = scaler.transform(X_test) 
```

```python
model = DecisionTreeClassifier(criterion='gini',max_depth=3)
model.fit(scaled_X_train, y_train)
y_pred = model.predict(scaled_X_test)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
plt.figure(figsize = (15, 15))
plot_tree(model, feature_names = pistachio.columns[:-1], 
          class_names = ['Kirmizi_Pistachio', 'Siit_Pistachio'], filled = True)
plt.show()
```
